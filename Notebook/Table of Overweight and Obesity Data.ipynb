```python
data = top_ten_obesity_13.sort_values('Data_value', ascending = True)
plt.figure(figsize = (8,4))
plt.bar('Region', 'Data_value', data = top_ten_obesity_13, color = '#15B5B0')
plt.xticks(rotation = 0)
plt.title('Top Five Obesity Rates 2013')
plt.annotate(text = '35.1%', xy = (-.15,31.5))
plt.annotate(text = '35.1%', xy = (.85,31.5))
plt.annotate(text = '34.6%', xy = (1.85,31))
plt.annotate(text = '33.7%', xy = (2.85,30))
plt.annotate(text = '33.2%', xy = (3.84,29.5));
```
