```python
from shapely.geometry import Point
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import folium
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
```

```python
overall_obesity_df = gpd.read_file('../cleaned_python_csv/obesity_overall.csv')
overall_obesity_df.head()
```

```python
overall_obesity_df[['lon', 'lat']] = overall_obesity_df['GeoLocation'].str.split(',', expand = True)
overall_obesity_df
```

```python
overall_obesity_df["lon"] = overall_obesity_df["lon"].str.replace("(","")
overall_obesity_df["lat"] = overall_obesity_df["lat"].str.replace(")","")
overall_obesity_df
```

```python
overall_obesity_df["lon"] = overall_obesity_df["lon"].astype(float)
overall_obesity_df["lat"] = overall_obesity_df["lat"].astype(float)
```

```python
overall_obesity_df["lon"] = pd.to_numeric(overall_obesity_df["lon"])
overall_obesity_df["lat"] = pd.to_numeric(overall_obesity_df["lat"])
overall_obesity_df
```

```python
overall_obesity_df = gpd.GeoDataFrame(
    overall_obesity_df, geometry=gpd.points_from_xy(overall_obesity_df.lon, overall_obesity_df.lat))
overall_obesity_df
```

```python
crs = {'init': 'epsg:4326'}
overall_obesity_df = gpd.GeoDataFrame(overall_obesity_df, crs=crs).set_geometry('geometry')
overall_obesity_df.head()
```

```python
ax = overall_obesity_df['Data_Value'].plot(
    color='white', edgecolor='black')

overall_obesity_df.plot(ax=ax, color='red')

plt.show()
```
