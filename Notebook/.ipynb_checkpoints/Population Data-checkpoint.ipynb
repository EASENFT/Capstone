```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
census_df = pd.read_csv('../data/Census_data.csv')
census_df
```

```python
census_df = census_df.replace(',','', regex=True)
census_df
#remove all commas
census_df['Region'] = census_df['Region'].str.strip('.')
census_df.head()
```

```python
census_df['Census'] = census_df['Census'].astype('int64')
census_df['Estimates Base'] = census_df['Estimates Base'].astype('int64')
census_df['2010'] = census_df['2010'].astype('int64')
census_df['2011'] = census_df['2011'].astype('int64')
census_df['2012'] = census_df['2012'].astype('int64')
census_df['2013'] = census_df['2013'].astype('int64')
census_df['2014'] = census_df['2014'].astype('int64')
census_df['2015'] = census_df['2015'].astype('int64')
census_df['2016'] = census_df['2016'].astype('int64')
census_df['2017'] = census_df['2017'].astype('int64')
census_df['2018'] = census_df['2018'].astype('int64')
census_df['2019'] = census_df['2019'].astype('int64')
#change datatypes
```
