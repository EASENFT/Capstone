```python
table_overweight_obesity.drop(['Locationabbr','Display_order', 'Data_value_unit', 'Data_value_type',
       'Data_Value_Footnote_Symbol', 'Data_Value_Footnote', 'DataSource',
       'ClassId', 'TopicId', 'LocationID', 'BreakoutID', 'BreakOutCategoryID',
       'QuestionID', 'ResponseID'], axis = 1)
```

```python
table_overweight_obesity.rename(columns = {'Locationdesc':'Region'}, inplace = True)
table_overweight_obesity
```

```python
table_overweight_obesity['Response'] = table_overweight_obesity['Response'].str.replace('bmi', 'BMI')
table_overweight_obesity.Response.value_counts()                                         
```

```python
race_ethinicity_df  = table_overweight_obesity[table_overweight_obesity['Break_Out_Category'].isin(['Race/Ethnicity'])]
race_ethinicity_df
```

```python
age_df = table_overweight_obesity[table_overweight_obesity['Break_Out_Category'].isin(['Age Group'])]
age_df.head()
```

```python
income_df = table_overweight_obesity[table_overweight_obesity['Break_Out_Category'].isin(['Household Income'])]
income_df.head()
```

```python
education_df = table_overweight_obesity[table_overweight_obesity['Break_Out_Category'].isin(['Education Attained'])]
education_df.head()
```

```python
overall_df = table_overweight_obesity[table_overweight_obesity['Break_Out_Category'].isin(['Overall'])]
overall_df.head()
```

```python
gender_df =table_overweight_obesity[table_overweight_obesity['Break_Out_Category'].isin(['Gender'])]
gender_df.head()
```

```python
gender_df.to_csv(r'../cleaned_python_csv/obesity_gender.csv')
income_df.to_csv(r'../cleaned_python_csv/obesity_income.csv')
race_ethinicity_df.to_csv(r'../cleaned_python_csv/obesity_race_ethinicity.csv')
overall_df.to_csv(r'../cleaned_python_csv/obesity_overall.csv')
education_df.to_csv(r'../cleaned_python_csv/obesity_education.csv')
age_df.to_csv(r'../cleaned_python_csv/obesity_age.csv')
```
