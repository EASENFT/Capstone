```python
median_income_2019.columns
```

```python
median_income_2019.rename(columns = {'Name':'Region'}, inplace = True)
median_income_2013.rename(columns = {'Name':'Region'}, inplace = True)
median_income_2016.rename(columns = {'Name':'Region'}, inplace = True)
median_income_2019
```

```python
data = median_income_2019.sort_values('2019')
plt.figure(figsize = (20,20))
plt.bar('Region', '2019', data = data, color = 'purple')
plt.xticks(rotation = 70)
plt.title('Median Income by State')
plt.show();
```

```python
data_2 = median_income_2016.sort_values('2016')
plt.figure(figsize = (20,20))
plt.bar('Region', '2016', data = data_2,)
plt.xticks(rotation = 70)
plt.title('Median Income by State')
plt.show();
```

```python
data_3 = median_income_2013.sort_values('2013')
plt.figure(figsize = (20,20))
plt.bar('Region', '2013', data = data_3, color = 'red')
plt.xticks(rotation = 70)
plt.title('Median Income by State')
plt.show();
```

```python
median_income_all_years = median_income_2019.merge(median_income_2016, on='Region', how='left').merge(median_income_2013, on='Region', how='left')
median_income_all_years
```

```python
fig, ax = plt.subplots(figsize=(25, 15))
x = np.arange(len(median_income_all_years.Region.unique()))

# Define bar width. We'll use this to offset the second bar.
bar_width = 0.4

# Note we add the `width` parameter now which sets the width of each bar.
b1 = ax.bar(x, median_income_all_years['2019'],
            width=bar_width)
# Same thing, but offset the x by the width of the bar.
b2 = ax.bar(x + bar_width, median_income_all_years['2016'],
            width=bar_width)
b3 = ax.bar(x + bar_width, median_income_all_years['2013'],
            width=bar_width)
ax.set_xticks(x + bar_width / 2)
ax.set_xticklabels(median_income_all_years.Region.unique(), rotation = 90 )

ax.set_xlabel('Region', labelpad=25)
ax.set_ylabel('# Median Income', labelpad=25)
ax.set_title('Median Income by State', pad=25)
```
